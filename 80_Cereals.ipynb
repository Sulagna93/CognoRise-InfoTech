```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(color_codes=True)
```

```python
# Load the Dataset
df = pd.read_csv('cereal.csv')
df
```

```python
#Checking for null values
null_counts = df.isnull().sum()
if null_counts.sum()>0:
    print(null_counts[null_counts>0])
```

```python
# Create a Bar plot
plt.figure(figsize=(5,4))
plt.bar(df['mfr'].value_counts().index, df['mfr'].value_counts().values)
plt.xlabel('Manufacturer')
plt.ylabel('Count')
plt.title('Number of Cereals by Manufacturer')
plt.show()
```

```python
# Create a Bar plot
plt.figure(figsize=(5,4))
plt.bar(df['type'].value_counts().index, df['type'].value_counts().values)
plt.xlabel('Type')
plt.ylabel('Count')
plt.title('Cereals by Cold or Hot')
plt.show()
```

```python
# Histogram
plt.figure(figsize=(5,4))
plt.hist(df['calories'], bins=10, color='salmon')
plt.title('Distribution of Calories per Serving')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box plot
plt.figure(figsize=(6,5))
sns.boxplot(x='sugars', y='mfr', data=df)
plt.xlabel('Manufacturer')
plt.ylabel('Grams of Sugar')
plt.title('Sugar Content by Manufacturer')
plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(5,4))
sns.scatterplot(x='calories', y='rating', data=df, color='green')
plt.xlabel('Calories')
plt.ylabel('Rating')
plt.title('Calories vs. Rating')
plt.show()
```

```python
# Pie plot
plt.figure(figsize=(5,4))
plt.pie(df['type'].value_counts(), autopct='%1.1f%%', colors=['lightcoral', 'lightskyblue'])
plt.title('Proportion of Hot and Cold Cereals')
plt.ylabel('')
plt.legend(labels=['Cold', 'Hot'])
plt.show()
```
